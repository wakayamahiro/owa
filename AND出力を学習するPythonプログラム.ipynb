##AND出力を学習するPythonプログラム

```python
import random
import numpy as np

#入力を0か1かランダムで選ぶ
x1 = [0, 1]

#入力をいれる所
x = []

#重みを２つ初期値で作る。
w = np.random.rand(2,)

#重みを入れる所
yoi_w1 = []
yoi_w2 = []


#重みを考える
for i in range(100):

#入力の値をランダムにxのリストに入れる。
  for i in range(2):
    x2 = random.choice(x1)
    x.append(x2)

#パーセプトロンの式
  a = x[0] * w[0] + x[1] * w[1]
  an = a
  #print('a',a)

  print('w0 = ',w[0])
  print('w1 = ',w[1])
  print('x0 = ',x[0])
  print('x1 = ',x[1])
  print('anser',a)

#出した数字と答えの誤差を求める
#(数字-答え)の二乗
  if x[0]==1:
    if x[1]==1:
      gosa = (an - 1)**2
    else:
      gosa = (an - 0)**2
  else:
    gosa = (an - 0)**2

  print('gosa',gosa)



#重みを調整する
#重みを変更した場合「重みの変更」が出力される
#重みの変更をすると[omo=1]になる。
  omo = 0
  if x[0]==1:
    if x[1]==1:

#[1][1]
      if gosa<=0.1:
        pass
      else:
        print('重みの変更')
        omo = 1
        w[0] = w[0] + 0.05*(1-a)*a*(1-a)*x[0]
        w[1] = w[1] + 0.05*(1-a)*a*(1-a)*x[1]

#[1][0]
    else:
      if gosa<=0.1:
        pass
      else:
        print('重みの変更')
        omo = 1
        w[0] = w[0] - 0.05*(0-a)*a*(1-a)*x[0]
        w[1] = w[1] - 0.05*(0-a)*a*(1-a)*x[1]

#[0][1]
  else:
    if x[1]==1:

      if gosa<=0.1:
        pass
      else:
        print('重みの変更')
        omo = 1
        w[0] = w[0] - 0.05*(0-a)*a*(1-a)*x[0]
        w[1] = w[1] - 0.05*(0-a)*a*(1-a)*x[1]

#[0][0]
    else:
      if gosa<=0.1:
        pass
      else:
        print('重みの変更')
        omo = 1
        w[0] = w[0] - 0.05*(0-a)*a*(1-a)*x[0]
        w[1] = w[1] - 0.05*(0-a)*a*(1-a)*x[1]
  

#0.5より大きかったら[1] 小さかったら[0]
#重みの変更をした場合、誤差が大きいので上の法則の逆で出力される
  if omo==1:
    if a <= 0.5:

      print('答え「１」')
    else:
      print('答え「０」')

  else:
    if a >= 0.5:
      print('答え「１」')
    else:
      print('答え「０」')

#入力をリセット
  x = []
#重み変更を行ったか？をリセット
  omo = 0

  print(':::::::::::::::::')
```
