##画像を塗り絵にするシステム

画像を白黒にする

outputに画像を保存

３．エッジを抽出する

```python
import numpy as np
import cv2

#1. 画像入力
#2. グレースケール化(グレースケールで入力)
img = cv2.imread('hall1.jpg', cv2.IMREAD_GRAYSCALE)

#3. 膨張処理
neiborhood = np.ones((5, 5), dtype=np.uint8)
kernel = np.ones((5,5),np.uint8)
dilated = cv2.dilate(img,kernel,iterations = 1)

#4. ２と３の差分取得
diff = cv2.absdiff(dilated, img)

#5. 白黒反転
result = 255 - diff
#result = diff

#画像出力
cv2.imwrite('output.jpg', result)
```

画像の表示

```python
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

img = Image.open('output.jpg')
plt.imshow(np.array(img))
```

線が薄い所と濃い所があるので、少しでも色があったらその部分をはっきり黒にする

atに保存する

```python
import cv2
import numpy as np

#画像の読み込み
img = cv2.imread("output.jpg")

#グレースケール変換
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

#閾値の設定
threshold_value = 224

#配列の作成（output用）
threshold_img = gray.copy()

#実装(numpy)
threshold_img[gray < threshold_value] = 0
threshold_img[gray >= threshold_value] = 255

#Output
cv2.imwrite("at1.jpg",threshold_img)
```

画像の表示

```python
img = Image.open('at1.jpg')
plt.imshow(np.array(img))
```
